```python
import numpy as np
import pandas as pd
from collections import defaultdict
import utils3 as ut
import random
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score
import logging
from tqdm import tqdm
import shap
from shap.plots import bar, waterfall, force
import warnings

warnings.filterwarnings('ignore')

logging.basicConfig(level=logging.DEBUG, format="[%(asctime)-15s] %(message)s", datefmt="%Y-%m-%d %H:%M:%S")
shap.initjs()
```

```python
# Récupération des données
# Initialisation d'un tableau
# Fonction de génération de données synthétiques
# Génération de lignes
# Entraînement d'un modèle de prédiction
# Visualisation des SHAP values
```

```python
# Récupération des données
```

```python
input =  {'A1': (119, 188), 'A2': (56, 262), 'A3': (67, 290), 'B1': (152, 322), 'B2': (147, 343), 'C0': (193.77578001057643, 41.32443151771548), 'C1': (184, 179), 'C2': (181, 224), 'C3': (177, 306), 'C4': (173, 334), 'C5': (172, 348), 'D1': (198, 324), 'D2': (199, 347), 'E1': (247.07615398014286, 197.09416477965442), 'E2': (298.9879115136736, 279.2535795157628), 'E3': (284.14189165650583, 305.4183591708761)}
Diam = 82
Long = 78
```

```python
# Initialisation d'un tableau
```

```python
tableau = {}
for i in input.keys():
    tableau[f"{i}_x"] = input[i][0]
    tableau[f"{i}_y"] = input[i][1]
    tableau["longueur_poro"] = ut.calcul_(Diam, Long, input['A1'], input['E1'], input['C1'], input['C2'], input['C3'], input['C4'], input['C5'], input['A2'], input['E2'], input['A3'], input['E3'], input['B1'], input['D1'])[0]
    tableau["largeur_poro"] = ut.calcul_(Diam, Long, input['A1'], input['E1'], input['C1'], input['C2'], input['C3'], input['C4'], input['C5'], input['A2'], input['E2'], input['A3'], input['E3'], input['B1'], input['D1'])[1]
data = pd.DataFrame(tableau, index=[0])
```

```python
data = data[['A1_x', 'A1_y', 'A2_x', 'A2_y', 'A3_x',
       'A3_y', 'B1_x', 'B1_y', 'B2_x', 'B2_y', 'C0_x', 'C0_y', 'C1_x', 'C1_y',
       'C2_x', 'C2_y', 'C3_x', 'C3_y', 'C4_x', 'C4_y', 'C5_x', 'C5_y', 'D1_x',
       'D1_y', 'D2_x', 'D2_y', 'E1_x', 'E1_y', 'E2_x', 'E2_y', 'E3_x', 'E3_y', 
       'longueur_poro', 'largeur_poro']]
data
```

```python
# Fonction de génération de données synthétiques sur un tuple
def synthetic(a):
    return a[0]+random.randint(-5, 5), a[1]+random.randint(-5, 5)
```

```python
# Génération de lignes
```

```python
for l in tqdm(range(10000)):
    input_s = {k : synthetic(input[k]) for k in input}
    longueur, largeur = ut.calcul_(Diam, Long, input_s['A1'], input_s['E1'], input_s['C1'], input_s['C2'], input_s['C3'], input_s['C4'], input_s['C5'], input_s['A2'], input_s['E2'], input_s['A3'], input_s['E3'], input_s['B1'], input_s['D1'])  
    ligne = dict(zip([f"{i}_x" for i in input_s.keys()], [input_s[i][0] for i in input_s.keys()]))|\
    dict(zip([f"{i}_y" for i in input_s.keys()], [input_s[i][1] for i in input_s.keys()]))|{'longueur_poro' : longueur, 'largeur_poro' : largeur}
    data = pd.concat([data, pd.DataFrame(ligne, index=[l])])
data = data.reset_index(drop=True)
data
```

```python
data = data[['C3_x', 'C2_y', 'C4_y', 'C3_y', 'C4_x', 'C2_x', 'E1_x', 'A1_x', 'A1_y', 'longueur_poro', 'largeur_poro']]
```

```python
# Entraînement d'un modèle de prédiction
model = LinearRegression()
model_longueur = model.fit(data.iloc[:,:-2], data.iloc[:,-2])
```

```python
y_pred = model_longueur.predict(data.iloc[:, :-2])
```

```python
r2_score(data.iloc[:,-2], y_pred)  
logging.debug (f"model {model} r2_score {r2_score(data.iloc[:,-2], y_pred)}")
```

```python
# Visualisation des SHAP values
```

```python
# Fits the explainer
explainer = shap.Explainer(model_longueur.predict, data.iloc[:100,:-2])
# Calculates the SHAP values - It takes some time
shap_values = explainer(data.iloc[:5,:-2])
```

```python
bar(shap_values)
```

```python
waterfall(shap_values[0])
```

```python
# La base value es tla valeur moyenne de toutes les valeurs et la value f(x) est la valeur prédite
force(shap_values[0])
```

```python
shap.summary_plot(shap_values, plot_type='violin')
```
