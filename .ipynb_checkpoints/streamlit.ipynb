```python
import glob

import streamlit as st
from PIL import Image, ImageDraw

from streamlit_image_coordinates import streamlit_image_coordinates

st.set_page_config(
    page_title="Streamlit Image Coordinates: Image Update",
    page_icon="üéØ",
    layout="wide",
)

"# :dart: Streamlit Image Coordinates: Image Update"

if "points" not in st.session_state:
    st.session_state["points"] = []


"## Click on image"


def get_ellipse_coords(point: tuple[int, int]) -> tuple[int, int, int, int]:
    center = point
    radius = 10
    return (
        center[0] - radius,
        center[1] - radius,
        center[0] + radius,
        center[1] + radius,
    )

# Liste des fichiers dans le dossier data
image_files = [f for f in glob.glob("data/*.jpg")]

# S√©lection de l'image
selected_image = st.sidebar.selectbox("S√©lectionnez une image:", image_files)

# Affichage de l'image et enregistrement du click
st.write("## image")

with st.echo("below"), Image.open(selected_image) as img:
    #draw = ImageDraw.Draw(img)
    draw = streamlit_image_coordinates(
        selected_image,
        width=500,
        key="jpg",
    )

    # Draw an ellipse at each coordinate in points
    for point in st.session_state["points"]:
        coords = get_ellipse_coords(point)
        draw.ellipse(coords, fill="red")

    value = streamlit_image_coordinates(img, key="pil")

    if value is not None:
        point = value["x"], value["y"]

        if point not in st.session_state["points"]:
            st.session_state["points"].append(point)
            st.rerun()
```

```python
import sympy as sp
import matplotlib.pyplot as plt

# D√©finir les points
x1, y1 = 1, 2
x2, y2 = 4, 6

# D√©finir les symboles
x, y = sp.symbols('x y')

# √âquation de la droite passant par deux points (x1, y1) et (x2, y2)
slope = (y2 - y1) / (x2 - x1)
intercept = y1 - slope * x1
line_eq = slope * x + intercept - y

# √âquation de la perpendiculaire passant par un point (x1, y1)
perpendicular_slope = -1 / slope
perpendicular_eq = perpendicular_slope * (x - x1) + y1 - y

# √âquation de la parall√®le passant par un point (x1, y1)
parallel_eq = slope * (x - x1) + y1 - y

# Trouver le point d'intersection entre la droite et la perpendiculaire
intersection = sp.solve([line_eq, perpendicular_eq], (x, y))

# Tracer les lignes
plt.figure(figsize=(10, 6))

# Tracer la droite
plt.plot([x1, x2], [y1, y2], label='Droite')

# Tracer la perpendiculaire
perp_x = [intersection[x], x1]
perp_y = [intersection[y], y1]
plt.plot(perp_x, perp_y, label='Perpendiculaire')

# Tracer la parall√®le
parallel_x = [x1 - 1, x1 + 1]
parallel_y = [slope * (xi - x1) + y1 for xi in parallel_x]
plt.plot(parallel_x, parallel_y, label='Parall√®le')

# Marquer les points
plt.scatter([x1, x2], [y1, y2], color='red')
plt.scatter([intersection[x]], [intersection[y]], color='blue')

# Ajouter des l√©gendes et des labels
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Droite, Perpendiculaire, Parall√®le et Intersection')
plt.grid(True)
plt.show()
```
